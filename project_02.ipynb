### Работа с остальными замечаниями:

```python
# проверю сколько объявлений с нулевым количеством комнат
display(data[data['rooms'] == 0]['rooms'].count())

# может это студии?
display(data[(data['rooms'] == 0) & (data['studio'] == True)]['rooms'].count())

# а сколько объявлений с заявленной жилой площадью менее 6 квадратных метров?
display(data[data['living_area'] < 6]['living_area'].count())

# сколько потолков ниже СНиП-овской нормы для жилых комнат в 2.5 метра?
display(data[data['ceiling_height'] < 2.5]['ceiling_height'].count())

# а пололков выше - ну пусть будет 5 метров, больше чем в элитных Московских пентхаузах
display(data[data['ceiling_height'] > 5]['ceiling_height'].count())
```

Теперь необходимо проверить таблицу на явные дубликаты. Также, в столбце с названиями населенных пунктов могут быть неявные дубликаты, способные повлиять на анализ.

```python
# проверим на явные дубликаты
display(data.duplicated().sum())

# явных дубликатов нет, что с названиями населенных пунктов? Проверю 30 самых публикуемых названий.
data['locality_name'].value_counts().head(30)
```

Выборка действительно по Санкт-Петербургу и окрестностям. Как минимум, дублируются деревня Кудрово и Кудрово. Необходимо оставить названия наиболее публикуемых населенных пунктов без их типа.

```python
# признаю, решение взял от другого студента в слаке, чуть изменил, но главное что мне оно понятно
# функция проверяет наличие ключевых слов в строке, возвращает эту же строку без ключевого слова + 1 символ за пробел

def delete_loc_type(loc_name):
    type_list = [
    'деревня',
    'посёлок',
    'поселок',
    'село'
    ]
    for i in type_list:
        if i in loc_name:
            return loc_name[len(i) + 1:]
        else:
            return loc_name
        
# применяю функцию к каждой строке
data['locality_name'] = data['locality_name'].apply(delete_loc_type, 1)

# проверяю
data['locality_name'].value_counts().head(30)
```

Предобработка закончена, можно переходить к расчетам.

## Расчёты и добавление результатов в таблицу

**Посчитайте и добавьте в таблицу:**  
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# подсчет цены квадратного метра это простое деление цены на площадь
data['square_price'] = data['last_price'] / data['total_area']

data['square_price'].head()
```

```python
# создание столбцов с днем недели, месяцем и годом публикации это простое применение dt методов
data['year'] = data['first_day_exposition'].dt.year
data['month'] = data['first_day_exposition'].dt.month
data['weekday'] = data['first_day_exposition'].dt.weekday

data[['year', 'month', 'weekday']].head(5)
```

```python
# для категоризации квартир по этажности создам функцию
# сравнение с единицей прежде сравнения с
def floors_group(index):
    if index['floor'] == 1:
        return 'первый'
    if index['floor'] == index['floors_total']:
        return 'последний'
    return 'другой'

# создаю столбец 'floors_group'
data['floors_group'] = data.apply(floors_group, axis=1)

# проверяю
data['floors_group'].value_counts()
```
