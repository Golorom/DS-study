### Описание техпроцесса

Как золото получают из руды?   
Изучите внимательно стадии процесса.  
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

![](https://pictures.s3.yandex.net/resources/viruchka_1576238830.jpg "Схема очистки золотосодержащей руды")

Опишем каждую стадию:
1. Флотация  
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.  
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка  
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

Технологический процесс:
- `Rougher feed` — исходное сырье
- `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- `Xanthate **`— ксантогенат (промотер, или активатор флотации);
- `Sulphate` — сульфат (на данном производстве сульфид натрия);
- `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

Параметры этапов:
- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи

Наименование признаков должно быть такое:  
`[этап].[тип_параметра].[название_параметра]`  
Пример: `rougher.input.feed_ag`

Возможные значения для блока [этап]:
- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

Возможные значения для блока [тип_параметра]:
- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

![](https://pictures.s3.yandex.net/resources/Recovery_1576238822.jpg "Формулы расчета эффективности очистки")

где:
- `C` — доля золота в концентрате после флотации/очистки;

```python
# на основе предоставленной формулы подготовлю функцию для оценки правильности расчета эффективности обогащения

def recovery(c, f, t):
    recovery = (c * (f - t)) / (f * (c - t)) * 100
    
    recovery[recovery<0] = np.nan
    recovery[recovery>100] = np.nan
    
    return recovery
```

```python
# применю ее к тестовой выборке для расчета rougher.output.recovery

true_recovery = recovery(
        gold_recovery_train['rougher.output.concentrate_au'],
        gold_recovery_train['rougher.input.feed_au'],
        gold_recovery_train['rougher.output.tail_au'])
```

```python
# сравню с исходными данными

mae = mean_absolute_error(
    gold_recovery_train['rougher.output.recovery'].fillna(0),
    true_recovery.fillna(0))

print('Средняя абсолютная ошибка:', mae)
```

### Подготовка данных

Обработаю пропущенные значения.

В данной ситуации можно использовать два способа работы с пропусками:

Первый:
- Заполнить пропуски методом 'ffill', так как данные между близкими датами отличаются минимально.

Второй:
- Избавиться от всех данных с пропусками.

Первый метод позволяет сохранить большее количество данных, но может возрасти их повторяемость, либо могут быть внесены искажения. Во втором методе теряется около 9% полного датасета, но данных должно быть достаточно для построения модели. 

Первый метод можно применить после завершения остальных этапов подготовки данных. Второй метод следует применять перед началом остальных этапов подготовки данных.

Выберу второй метод так как данных для подготовки модели должно быть достаточно и доля потери не критична, а чистота данных важна.

```python
gold_recovery_train = gold_recovery_train.dropna()
gold_recovery_test = gold_recovery_test.dropna()
gold_recovery_full = gold_recovery_full.dropna()
```

```python
display(gold_recovery_train.isna().sum().sum())
display(gold_recovery_test.isna().sum().sum())
gold_recovery_full.isna().sum().sum()
```

Избавлюсь от признаков с корреяцией выше 0.95.

```python
# список признаков для расчета корреляции
corr_cols = ['final.output.recovery', 'rougher.output.recovery'] + list(gold_recovery_test.columns)
```

```python
# на основе списка получу матрицу корреляции
corr_matrix = gold_recovery_train[corr_cols].corr().abs()
corr_matrix
```

```python
# избавлюсь от дублирующихся данных и диагонали
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
upper_tri
```

```python
# определю признаки с высокой корреляцией
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
to_drop
```

```python
# избавлюсь от этих признаков в обучающей и тестовой выборках
# удалю из обучающей выборки признаки отсутствующие в тестовой выборке

gold_recovery_test = gold_recovery_test.drop(to_drop, axis = 1)

gold_recovery_train = gold_recovery_train[gold_recovery_test.columns]

missing_cols = gold_recovery_train.columns.difference(
    gold_recovery_test.columns)

print(missing_cols)
```

```python
# количество признаков должно стать одинаковым
display(gold_recovery_test.shape)
gold_recovery_train.shape
```

```python
# использую временные метки в качестве индекса для установления взаимосвязи признаков между выборками и полным датасетом
gold_recovery_test['date'] = pd.to_datetime(gold_recovery_test['date'])
gold_recovery_test.set_index('date', inplace=True, verify_integrity=True)

gold_recovery_train['date'] = pd.to_datetime(gold_recovery_train['date'])
gold_recovery_train.set_index('date', inplace=True, verify_integrity=True)

gold_recovery_full['date'] = pd.to_datetime(gold_recovery_full['date'])
gold_recovery_full.set_index('date', inplace=True, verify_integrity=True)
```

### Функция высичления итоговой sMAPE

Итоговая sMAPE для текущей задачи это сумма 25% доли от промежуточного целевого признака и 75% от финального целелвого признака

```python
def smape_calc(target, prediction):
    num = np.abs(target - prediction)
    denom = (np.abs(target) + np.abs(prediction)) / 2 
    
    return 1 / len(target) * np.sum(num / denom) * 100 
    
    
def total_smape_calc(target, prediction):
    target_rougher = target.iloc[:, 0]
    target_final = target.iloc[:, 1]
    
    prediction_rougher = prediction[:, 0]
    prediction_final = prediction[:, 1]
    
    return 0.25 * smape_calc(target_rougher, prediction_rougher) + 0.75 * smape_calc(target_final, prediction_final)
```

```python
# scorer позволит использовать итоговую sMAPE для оценки качества моделей уже на этапе кросс-валидации
# (вообще отилчная штука, жаль что в программе курса не упоминается)
smape_scorer = make_scorer(total_smape_calc, greater_is_better=False)
```

### Дерево решений

```python

dtr = DecisionTreeRegressor(random_state=STATE)

params_tree = [{
    'max_depth': range(2, 7),
    'min_samples_split': range(2, 4),
    'min_samples_leaf': range(1, 3)
}]

model_tree = GridSearchCV(dtr,
                      param_grid=params_tree,
                      scoring=smape_scorer)

model_tree.fit(features_train, target_train)

# model_tree.best_score_, так как при параметре "greater_is_better=False" возвращается значение с обратным знаком
print('Лучшее значение sMAPE для Дерева Решений:', -model_tree.best_score_)
'Оптимальные параметры модели:', model_tree.best_params_
```

### Случайный лес

```python
rfr = RandomForestRegressor(random_state=STATE)

params_forest = [{'n_estimators': range(10, 41, 10),
                 'max_depth': range(2, 4)}]

model_forest = GridSearchCV(rfr, params_forest, cv=5, scoring=smape_scorer)
model_forest.fit(features_train, target_train)

print('Лучшее значение sMAPE для Случайного Леса:', -model_forest.best_score_)
'Оптимальные параметры модели:', model_forest.best_params_
```

### Линейная регрессия

Наилучший результат получен на модели Случайного Леса. Проверю модель на тестовой выборке и сравню результат с отработкой глупой модели.
